# Program to count the frequency of each character in a word

The counter function from collections provide an easy way to caluclate the frequency of each word in a string

First part is to import the correct module from the package

```python
from collections import Counter
```

<div class="alert alert-block alert-info">
    Function to calculate the frequencies of each letter in a word and print them: </b>:</b></div>

```python
def freq_calc(word):
    frequency = Counter(word.lower())
    print(frequency)
    return "Task complete"
```

    Function to input the string from user </div>

```python
def get_Word():
    word = input("Enter the word to calculate the character frequency: ")
    print()
    return word
```
