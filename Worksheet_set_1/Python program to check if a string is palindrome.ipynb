# Palindrome

## What is a palindrome?

A palindrome string is one which reads the exact same from either side.
Example:<br> 
The String "12321" is palindrome.<br>
Similarly, the string "Radar" is palindrome.

<div class="alert alert-block alert-info">
    Function to find out if the String is <b> Palindrome </b>:</b></div>

Python string slicing allows an easy way to check if a string is palindrome or not

```python
def check_palindrome(word):
    if word.lower().strip() == word.lower().strip()[::-1]:
        return "Palindrome"
    else:
        return "Not Palindrome"
```

<div class="alert alert-block alert-info">
    Function to print results </div>

```python
def get_result(word):
    if word == "":
        return "The string is empty."
    else: 
        result = check_palindrome(word)
        print('The word "'+word+'" is '+result)
        return "Task Complete"
```

<div class="alert alert-block alert-info">
    Function to input the string from user </div>

```python
def get_String():
    word = input("Enter a string to check for palindrome: ")
    return word
```
