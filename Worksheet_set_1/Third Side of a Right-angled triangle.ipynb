# Find the third side of a Right-angled triangle from two given sides

Finding the sides of a triangle comes down to Pythagoras theorem<br>
Theorem: a^2 = b^2 + c^2<br>
We need to figure out which sides of the triangle is the user giving input. A hypotenuse, a base and a height are the 3 sides of a triangle.

Math function used to find out the square root

```python
import math
import sys
```

<div class="alert alert-block alert-info">
    Function to Calculate the third side of a Right-hand triangle </div>

```python
def get_third(hypotenuse,base,height):
    if hypotenuse == "?":
        hypotenuse = math.sqrt(base**2+height**2)
        return str(hypotenuse)
    elif base == "?":
        base = math.sqrt(hypotenuse**2-height**2)
        return str(base)
    elif height == "?":
        height = math.sqrt(hypotenuse**2-base**2)
        return str(height)
    else: 
        return ""
```

<div class="alert alert-block alert-info">
    Function to print results </div>

```python
def get_result(tup):
    h = tup[0]
    b = tup[1]
    he = tup[2]
    print()
    if h == "?":
        print("The hypotenuse of a triangle with base: "+str(b)+" and height: "+str(he)+" is: "+ get_third(h,b,he))
    elif b == "?":
        print("The base of a triangle with hypotenuse: "+str(h)+" and height: "+str(he)+" is: "+ get_third(h,b,he))
    elif he == "?":
        print("The height of a triangle with hypotenuse: "+str(h)+" and base: "+str(b)+" is: "+ get_third(h,b,he))
    else:
        print()
        return "Task Complete"
```

<div class="alert alert-block alert-info">
    Function to input the 2 sides from user and figure out what side the user wants to know? </div>

```python
def get_inputs():
    print("Please specify which side of the triangle do you want to find: H/B/He")
    side = input()
    if side.lower() == "h":
        H = "?"
        B  = int(input("Please enter the Base: "))
        He = int(input("Please enter the height: "))
        return H,B,He
    elif side.lower() == "b":
        H  = int(input("Please enter the Hypotenuse: "))
        B  = "?"
        He = int(input("Please enter the height: "))
        return H,B,He
    elif side.lower() == "he":
        H  = int(input("Please enter the Hypotenuse: "))
        B  = int(input("Please enter the Base: "))
        He = "?"
        return H,B,He
    else:
        raise SystemExit
```
