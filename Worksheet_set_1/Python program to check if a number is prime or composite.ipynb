# Prime or Composite

### What is a prime number?

A prime number is a number that has two factors, the number: 1 and itself. 
Example: The number 17 has two factors - 1 and 17 itself.

Conversely, a composite number is one which has more than 2 factors.
Example: The number 18 has the following factors - 1, 2, 3, 6, 9 and 18.

<div class="alert alert-block alert-info">
Function to find out if the number is <b>Prime</b> or <b>Composite :</b></div>

The concept behind prime numbers is if a number has a factor other than 1 and itself.

Example: 

<b>Case 1</b>:                5 has 2 factors - 1 and 5.<br>
&emsp;&emsp;&emsp;&nbsp;&nbsp;If we run a loop from 2 to 4, no number should be able to divide 5.<br>
&emsp;&emsp;&emsp;&nbsp;&nbsp;Which in this case is true and hence, 5 is a Prime number.
         
<b>Case 2</b>:18 has the following factors - 1, 2, 3, 6, 9 and 18.<br>
&emsp;&emsp;&emsp;&nbsp;&nbsp;If we run a loop from 2 - 17, more numbers are able to divide 18.<br>
&emsp;&emsp;&emsp;&nbsp;&nbsp;Which is why 18 is a Composite number.

```python
def check_prime_or_composite(number):
    number = int(number)
    if number == 1 or number == 2:
        return "Prime"
    for i in range(2, number-1):
        if number%i == 0:
            return "Composite"
        else:
            return "Prime"        
```

<div class="alert alert-block alert-info">
Function to print the result <b>Prime</b> or <b>Composite :</b></div>

```python
def get_result(number):
    print()
    result = check_prime_or_composite(number)
    print("The number "+number+" is : "+ result )
    return "Task Complete"
```

```python
def get_number():
    number = input("Enter a number to check for Prime or Composite:")
    return number
```
