# Python program to find the Factorial of a number

### What is factorial?
The factorial of a number is the product of all the numbers from  1 to the number.
Factorial symbol (!)
Example:

5! = 1\*2\*3\*4\*5 = 120

<div class="alert alert-block alert-info">
 1. Function to find out the <b>Factorial</b> using <b>For Loop:</b></div>

The long process to obtain a factorial is to run a loop from 1 to the number obtained as the paramter

```python
def get_factorial_loop(n):
    factorial = 1
    for i in range(1, n + 1):
        factorial = factorial * i
    return factorial
```

<div class="alert alert-block alert-info">
 2. Function to find out the <b>Factorial</b> using <b>Recursion:</b></div>

The process is to set a base return condition and then recrusively call the same function until the exit condition is matched.

```python
def get_factorial_recursion(n):
    if n==1:
        return 1
    else:
        return n*get_factorial_recursion(n-1)
```

<div class="alert alert-block alert-info">
    Getting the input for a number and finding out the factorial using:
    1. For loop  using
    2. Recursion</div>

```python
def print_factorials(number):
    #Factorial through loop
    print("Factorial of " + str(number)+ " through For loop : "+ str(get_factorial_loop(number)))
      
    #Factorial through recursion
    print("Factorial of " + str(number)+ " through Recursion: "+ str(get_factorial_recursion(number)))
    
    print()
    return "Task complete"
```
